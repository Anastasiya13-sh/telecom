##  2 Подготовка данных 

<a id='Замена типа данных'></a>
### 2.1 Замена типа данных  

```python
#замена типа данных в таблице с данными прользователя
us['reg_date'] = us['reg_date'].astype('datetime64')
us['churn_date'] = us['churn_date'].astype('datetime64')

#замена типа данных в таблице с данными по звонкам
cal['call_date'] = cal['call_date'].astype('datetime64')

#замена типа данных в таблице с данными по интернету
inter['session_date'] = inter['session_date'].astype('datetime64')

#замена типа данных в таблице с данными по сообщениям
mes['message_date'] = mes['message_date'].astype('datetime64')
```

<a id='Исправление ошибок в данных'></a>
###  2.2 Исправление ошибок в данных

```python
#округлим минуты звонков, так как та тариф говорит нам, что проговорив 1 секунду, будет защитано, как 1 минута 
cal['duration'] = np.ceil(cal['duration'])

# добавим месяц отдельно в таблицу
cal['month'] = mes['message_date'].dt.month
cal.head(10)

#округлим трафик использованного интернета
inter['mb_used']= np.ceil(inter['mb_used'])
```

<a id='Подсчеты для каждого пользователя'></a>
###  2.3 Подсчеты для каждого пользователя 

```python
# выведим в таблице отдельно месяца, чтобы по ним  сделать подсчет  (в таблице сообщения)
mes['month'] = mes['message_date'].dt.month
mes.info()
mes.head(20)

# выведим в таблице отдельно месяца, чтобы по ним  сделать подсчет (в таблице интернет)
inter['month'] = inter['session_date'].dt.month
inter.head(20)
```

```python
#посчитаем общую длительность звонков каждого пользователя в месяц
total_calls_pivot = cal.pivot_table(index = ('user_id', 'month'), values = 'duration', aggfunc = 'sum').reset_index()
total_calls_pivot.head()
```

```python
#посчитаем количество отправленных сообщений по месяцам
total_message_pivot = mes.pivot_table(index = ('user_id', 'month'), values = 'id', aggfunc = 'count').reset_index()
total_message_pivot.head()
```

```python
#посчитаем объем  исрасходованного интернет трафика 
total_internet_pivot = inter.pivot_table(index = ('user_id', 'month'), values = 'mb_used', aggfunc = 'sum').reset_index()
total_internet_pivot.head()
```

```python
#соединим полученные данные в единую таблицу.
df = total_calls_pivot.merge(total_message_pivot, how='left', on=['user_id', 'month'])
df = df.merge(total_internet_pivot, how='left', on=['user_id', 'month'])
df = df.merge(us, how='left', on=['user_id'])
df = df.merge(tr, how='left', on=['tariff_name'])
display(df)
```

```python
#считаем помесячную выручку  с каждого пользователя 
df['balance_cal'] = df['duration'] - df['minutes_included']
df['balance_mes'] = df['id'] - df['messages_included'] 
df['balance_inter'] = df['mb_used'] - df['mb_per_month_included'] 
display(df)
```

```python
#заполним нулями значения значения, там где их не хватает
df.info()
df['balance_mes'] = df['balance_mes'].fillna(0)
df['balance_inter'] = df['balance_inter'].fillna(0)   
display(df)
```

```python
#сделаем отрицательные значения нулями
df['balance_mes'] = df['balance_mes'].clip(lower=0)
df['balance_cal'] = df['balance_cal'].clip(lower=0)
df['balance_inter'] = df['balance_inter'].clip(lower=0)
display(df)
```

```python
#переведем перепрасход интернета в ГБ и округлим 
df['overrun_balance_inter_gb'] = df['balance_inter'] / 1024
df['overrun_balance_inter_gb'] = np.ceil(df['overrun_balance_inter_gb'])

# узнаем прибыль за перерасход по звонкам для каждого пользователя 
df['profit_balance_cal'] = df['balance_cal'] * df['rub_per_minute']

# узнаем прибыль за перерасход по смс для каждого пользователя
df['profit_balance_mes'] = df['balance_mes'] * df['rub_per_message']

# узнаем прибыль за перерасход по интернету для каждого пользователя 
df['profit_balance_inter_gb'] = df['overrun_balance_inter_gb'] * df['rub_per_gb']

# сделаем отдельный столбец с общей суммой по всем перерасходом и абоненской платой за тариф 
df['total_profit'] = df['profit_balance_inter_gb'] + df['profit_balance_mes'] + df['profit_balance_cal'] + df['rub_monthly_fee']

display(df)
```

```python
#пользователь мог звонить, но не писать или не пользоваться интернетом. 
#Тогда нужно заполнить неизвестные данные нулями, чтобы в дальнейшнм это не мешало нам работать 

df['duration'] = df['duration'].fillna(0)
df['id'] = df['id'].fillna(0)
df['mb_used'] = df['mb_used'].fillna(0)
display(df)
```

#### Вывод

Мы подготовили данные для дальнейшего анализа. Соединили все таблицы и полученные данные в одну, чтобы в дальнейшем было легче анализировать и работать с данными

<a id='Проверка гипотез'></a>
##  4 Проверка гипотез

<a id='Первая гипотеза'></a>
###  4.1 Первая гипотеза 

Первая наша гипотеза звучит так 'средняя выручка пользователей таривов "Ультра" и "Смарт" различается'. Мы можем сказать, что  это двусторонняя гипотеза.  У нас есть две средние выручки тарифов, предположим, что они равны и возьмем это за нулевую гипотезу. альтернативной же  гипотезой будт то, что они не равны. Таким образом наша нулевая гипотеза звучит так 'средние значения по выручке равны'. 
Наши выборки независимы изза этого используем scipy.stats.ttest_ind. В аргументе equal_var укажем False, так как мы указали, что среднее значение по выборке не равны. 

```python
#проверим равны ли наши дисперсии выборок или нет
df_smart_var = np.var(df_smart['total_profit'], ddof=1)
df_ultra_var = np.var(df_ultra['total_profit'], ddof=1)
df_var = (df_smart_var / df_ultra_var)*100
#print(df_var)
if df_var < 5:
    print('дисперсии равны')
else:
    print('дисперсии не равны')
```

```python
#так как дисперсии не равны, то equal_var= False

from scipy import stats as st
alpha = .05
results = st.stats.ttest_ind(df_smart['total_profit'], df_ultra['total_profit'], equal_var= False)
print('p-value:', results.pvalue)
if (results.pvalue < alpha):
    print('отвергаем нулевую гипотезу')
else:
     print('не получилось отвергнуть нулевую гипотезу')
```

<a id='Вторая гипотеза'></a>
###  4.2 Вторая гипотеза 

Вторая гипотеза, которуют нам нужно проверить "отличается ли средняя выручка пользователей из Москвы  от выручки пользователей других регионов". Для этого сделаем две выборки, пользователи из Москвы и остальные пользователи. Наша нулевая гипотеза  будет звучить так 'средние значения по выручке Москвы равна выручке других регионов'. Альтернативная гипотеза будет звучать, что средние значени не равны. 
Наши выборки независимы изза этого используем scipy.stats.ttest_ind. 

```python
#сделаем новые выборки 
df_moscow = df.query('city == "Москва"')
df_another = df.query('city != "Москва"')

df_moscow_var = np.var(df_moscow['total_profit'], ddof=1)
df_another_var = np.var(df_another['total_profit'], ddof=1)
df_var = (df_moscow_var / df_another_var)*100
if df_var < 5:
    print('дисперсии равны')
else:
    print('дисперсии не равны')
```

```python
from scipy import stats as st
alpha = .05
results = st.stats.ttest_ind(df_moscow['total_profit'], df_another['total_profit'], equal_var=False)
print('p-value:', results.pvalue)
if (results.pvalue < alpha):
    print('отвергаем нулевую гипотезу')
else:
     print('не получилось отвергнуть нулевую гипотезу')
```
